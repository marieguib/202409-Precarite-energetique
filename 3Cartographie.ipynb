```python
# Import du fichier de résultats

import pandas as pd

# Import du fichier de résultats

df_resultats = pd.read_csv("data/resultats_proba_communes.csv")
df_resultats
```

```python
import requests

# https://geo.api.gouv.fr/decoupage-administratif/communes
url_api = "https://geo.api.gouv.fr/communes?code={}&fields=code,contour&format=json"

# Récupération des données géographiques pour chaque commune en utilisant le lien api en parrallélisant les requêtes

for code_insee in df_resultats['Code INSEE']:

    url = url_api.format(code_insee)
    response = requests.get(url)
    perimetre_communes = response.json()
    # # ajout de la colonne 'contour' dans le dataframe
    print(code_insee)
    print(perimetre_communes)
    if perimetre_communes :
        df_resultats.loc[df_resultats['Code INSEE'] == code_insee, 'contour'] = str(perimetre_communes[0]['contour'])
    else : 
        df_resultats.loc[df_resultats['Code INSEE'] == code_insee, 'contour'] = pd.NA
```

```python
# Export du dataframe avec les données géographiques

df_resultats.to_csv("data/resultats_proba_communes_contour.csv", index=False)
```
